Створіть дві матриці розміром 9x9, заповніть їх випадковими цілими числами в

Завдання 14
Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні
від 1 до 5та знайдіть середнє геометричне значення елементів матриці.

```python
# Створюємо матрицю розміром 3x3 з випадковими цілими числами від 1 до 5
matrix = np.random.randint(1, 6, size=(3, 3))

# Розгортаємо матрицю в одномірний масив
matrix_flat = matrix.flatten()

# Обчислюємо добуток всіх елементів
product = np.prod(matrix_flat)

# Обчислюємо середнє геометричне значення
geo_mean = product ** (1 / len(matrix_flat))

print("Матриця:")
print(matrix)
print("\nСереднє геометричне значення елементів матриці:")
print(geo_mean)
```

Завдання 15
Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в
діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить найменше за модулем
ціле значення в кожному рядку матриці.

```python
# Створюємо матрицю розміром 3x3 з випадковими дійсними числами від -2 до 2
matrix = np.random.uniform(-2, 2, size=(3, 3))

# Округлюємо матрицю до найближчих цілих чисел
matrix_rounded = np.round(matrix).astype(int)

# Знаходимо найменше за модулем ціле число в кожному рядку
min_abs_in_rows = np.sign(matrix_rounded) * np.min(np.abs(matrix_rounded), axis=1)

# Створюємо вектор-стовпчик
result_vector = min_abs_in_rows.reshape(-1, 1)

print("Матриця:")
print(matrix)
print("\nОкруглена матриця:")
print(matrix_rounded)
print("\nВектор-стовпчик з найменшими за модулем цілими числами в кожному рядку:")
print(result_vector)
```
